```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import pandas as pd
from tqdm.auto import tqdm
import torch

tqdm.pandas()
```

```python
test_df = pd.read_excel("validation_dataset.xlsx")
categories = test_df['label'].unique()
```

```python
def summarize_text(sentence):
    prompt = " Summarize the following in as few words as possible: " + sentence 
    return prompt

test_df['summarize_prompts'] = test_df['text'].progress_apply(summarize_text)
```

```python
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
input_ids = tokenizer.encode(test_df.loc[0, 'summarize_prompts'], return_tensors="pt", truncation=True, max_length=1024)
attention_mask = torch.ones(input_ids.shape, dtype=torch.long)
model = AutoModelForCausalLM.from_pretrained(model_name, pad_token_id=tokenizer.eos_token_id)
```

```python
for i in tqdm(range(len(test_df))):
    prompt = test_df.loc[i, 'summarize_prompts']
    input_ids = tokenizer.encode(prompt, return_tensors="pt", truncation=True, max_length = 1024)
    output = model.generate(input_ids, max_length = 1024, )
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    #print(f"Prompt: {prompt}")
    print(f"Generated Text: {generated_text}\n")
```
